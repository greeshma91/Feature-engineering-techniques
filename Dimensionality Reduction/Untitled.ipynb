```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
i=Image.open('pink.jpg')
img=i.convert('LA')
plt.figure(figsize=(16,9))
plt.imshow(img)
img.size
```

```python
# convert this image to equivalent numpy matrix
imgmat = np.array(list(img.getdata(band=0)),float)
imgmat.shape = (img.size[1],img.size[0])
imgmat = np.asmatrix(imgmat)
plt.figure(figsize=(16,9))
plt.imshow(imgmat,cmap='gray')
```

```python
#create U,S,V and svd and check shapes of it
U,S,V = np.linalg.svd(imgmat)
print(U.shape,S.shape,V.shape)
```

```python
# create reconstructed matrix and image of it
reconstimg = np.asmatrix(U[:,:1]) * np.diag(S[:1]) * np.asmatrix(V[:1,:])
plt.figure(figsize=(16,9))
plt.imshow(reconstimg,cmap='gray')
```

```python
# check for different n components
for i in range (5,51,5) :
    reconstimg = np.asmatrix(U[:,:i]) * np.diag(S[:i]) * np.asmatrix(V[:i,:])
    plt.figure(figsize=(16,9))
    plt.imshow(reconstimg,cmap='gray')
    plt.title('n=%s'%i)
```
