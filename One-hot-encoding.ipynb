```python
# OHE - is used to sort and order categorical variables based on their presence as 1 or 0 .
# eg. Here in the titanic dataset, lets see how we deal with 2 categorical variables - sex,embarked.
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
# Treating var sex :

data = pd.get_dummies(titanic['Sex'])
titanic = pd.concat([data,titanic],axis=1)
titanic = titanic.drop(['Sex'],axis = 1)
titanic.head()
```

```python
# Treating var Embarked : 

# no of unique values in embarked
titanic.Embarked.unique()
```

```python
data1 = pd.get_dummies(titanic['Embarked'])
titanic = pd.concat([data1,titanic],axis = 1)
titanic = titanic.drop(['Embarked'],axis = 1)
titanic.head()
```
